#Criando a tabela Relatorio
>Colocando os dados dentro da tabela




```python
import pandas as pd
import numpy as np


Relatorio = pd.DataFrame({'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
                      'Limpeza': [100, 0, 100, 0, 100, 100, 0],
                      'Comida': [221.60, 375.31, 412, 495.20, 441.53, 245, 164],
                      'Transporte': [150, 100, 125, 300, 275, 525, 75],
                      'Outros': [0, 0, 2310, 500, 0, 0, 820],
                      'Ganhos': [2200, 2420.50, 3391, 5322, 4898.50, 4200, 3893]})
Relatorio
```

#Subtraindo 7% de imposto dos ganhos

```python
Relatorio['Ganhos'] = Relatorio['Ganhos'] * 0.93

Relatorio['Ganhos']
```

#Soma total dos ganhos da semana

```python
soma_semana = Relatorio['Ganhos'].sum()

print("Soma dos ganhos da semana: %.2f"% soma_semana)
```

#Media semanal dos ganhos

```python
media_semana = Relatorio['Ganhos'].mean()

print("Media dos ganhos da semana: %.2f"% media_semana)
```

#Soma das despesas
>Ordenadas por categoria

```python
print('Soma dos gastos da limpeza:',Relatorio['Limpeza'].sum())

print('Soma dos gastos da comida:',Relatorio['Comida'].sum())

print('Soma dos gastos do transporte:',Relatorio['Transporte'].sum())

print('Soma dos gastos de outros:',Relatorio['Outros'].sum())
```

#Media semanal de todas as despesas

```python
mediasemanal = Relatorio['Limpeza'].sum() + Relatorio['Comida'].sum() + Relatorio['Transporte'].sum() + Relatorio['Outros'].sum() 

print('A media semanal de despesas está semana foi de:{:.2f}'.format(mediasemanal/4))
```

#Lucro diario
>Dia mais lucrativo e o lucro total da semana
>>Mostrando o resultado final da tabela

```python
Relatorio['Lucro'] = Relatorio['Ganhos'] - (Relatorio['Limpeza'] + Relatorio['Comida'] + Relatorio['Transporte'] + Relatorio['Outros'])

dia_mais_lucrativo = Relatorio['dia'][Relatorio['Lucro'].idxmax()]

lucro_total = Relatorio['Lucro'].sum()

print(f"Dia mais lucrativo: {dia_mais_lucrativo} e seu lucro foi de:{Relatorio['Lucro'].max():.2f}")
print(f"Lucro total da semana foi:{lucro_total:.2f}")

Relatorio
```
